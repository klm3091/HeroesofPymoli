```python
import pandas as pd
import numpy as np
```

```python
#Pull file
file = "Resources/purchase-data.csv"
purchase_data = pd.read_csv(file)
heroes_df = pd.DataFrame(purchase_data)
```

```python
#Calculate total players
total_players = len(heroes_df["SN"].unique())
total_players_df = pd.DataFrame({"Total Players":[total_players]})
total_players_df
```

```python
#Purchasing Analysis

#Find number of unique items
unique_items = heroes_df["Item ID"].nunique()
unique_items

#Calculate average price
average_price = heroes_df["Price"].mean()
average_price

#Total the number of purchases
total_purchases = heroes_df["Item ID"].count()
total_purchases

#Calculate the total revenue
total_price = heroes_df["Price"].sum()
total_price

#Summary Table
summary = pd.DataFrame({"Number of Unique Items ":[unique_items],
                            "Average Price " : [average_price],
                            "Number of Purchases ": [total_purchases],
                            "Total Revenue " : [total_price]})
summary
```

```python
#Gender Demographics

#Group by Gender & SN
grouped_gender = heroes_df.groupby(["SN","Gender"])
grouped_df = pd.DataFrame(grouped_gender.size())

#Calculate totals
gender_df = pd.DataFrame(grouped_df.groupby(["Gender"]).count())
gender_df.columns= ["Total"]
female = gender_df["Total"].loc["Female"]
male = gender_df["Total"].loc["Male"]
other = gender_df["Total"].loc["Other / Non-Disclosed"]

#Calculate percentages
gender_df["Percentage of Players"] = round(100*(gender_df["Total"]/gender_df["Total"].sum()),2)
gender_df = gender_df[["Percentage of Players","Total"]]

gender_df
```

```python
#Purchasing Analysis by Gender

#Percentage and count of females
female_purchases = len(heroes_df[heroes_df["Gender"] == "Female"])
female_total = '${:,.2f}'.format(round(heroes_df[heroes_df["Gender"] == "Female"]["Price"].sum(),2))
female_avg = '${:,.2f}'.format(round(heroes_df[heroes_df["Gender"] == "Female"]["Price"].mean(),2))
female_tot = '${:,.2f}'.format(round(heroes_df[heroes_df["Gender"] == "Female"]["Price"].sum(),2)/female)

#Percentage and count of males
male_purchases = len(heroes_df[heroes_df["Gender"] == "Male"])
male_total = '${:,.2f}'.format(round(heroes_df[heroes_df["Gender"] == "Male"]["Price"].sum(),2))
male_avg = '${:,.2f}'.format(round(heroes_df[heroes_df["Gender"] == "Male"]["Price"].mean(),2))
male_tot = '${:,.2f}'.format(round(heroes_df[heroes_df["Gender"] == "Male"]["Price"].sum(),2)/male)

#Percentage and count of other/non-disclosed
other_purchases = len(heroes_df[heroes_df["Gender"] == "Other / Non-Disclosed"])
other_total = '${:,.2f}'.format(round(heroes_df[heroes_df["Gender"] == "Other / Non-Disclosed"]["Price"].sum(),2))
other_avg = '${:,.2f}'.format(round(heroes_df[heroes_df["Gender"] == "Other / Non-Disclosed"]["Price"].mean(),2))
other_tot = '${:,.2f}'.format(round(heroes_df[heroes_df["Gender"] == "Other / Non-Disclosed"]["Price"].sum(),2)/other)

#Display analysis
gender_purchase_analysis_df = pd.DataFrame({"Gender":["Male", "Female","Other / Non-Disclosed"],
                                            "Purchase Count":[male_purchases,female_purchases,other_purchases],
                                            "Average Purchase Price":[male_avg,female_avg,other_avg],
                                            "Total Purchase Value":[male_total,female_total,other_total],
                                            "Average Total Per Person":[male_tot,female_tot,other_tot]},
                                            columns = ["Gender","Purchase Count",
                                                      "Average Purchase Price","Total Purchase Value",
                                                      "Average Total Per Person"])

gender_purchase_analysis_df.set_index("Gender", inplace=True) 
gender_purchase_analysis_df
```

```python
#Age Demographics

bins = [0, 9, 14, 19, 24,29,34,39,100]
group_names = ['<10', '10-14', '15-19', '20-24','25-29','30-34','35-39','40+']
pd.cut(heroes_df["Age"], bins, labels=group_names)
heroes_df["Age Group"] = pd.cut(heroes_df["Age"], bins, labels=group_names)

groupby_sn_agegrp = heroes_df.groupby(["SN","Age Group"])

                                                              
age_payers_df= pd.DataFrame(groupby_sn_agegrp.size())

purchase_agegrp_df = pd.DataFrame(age_payers_df.groupby(["Age Group"]).count())
purchase_agegrp_df.columns= ["Total Count"]
purchase_agegrp_df["Percentage of Players"] =round(100 *purchase_agegrp_df["Total Count"]/purchase_agegrp_df["Total Count"].sum(),2)

purchase_agegrp_df
```

```python
#Purchasing Analysis (Age)

age_count_df = pd.DataFrame(heroes_df.groupby(["Age Group"]).count())
age_sum_df = pd.DataFrame(heroes_df.groupby(["Age Group"]).sum())
purchase_agegrp_df["Purchase Count"] =  age_count_df["Item ID"]
purchase_agegrp_df["Average Purchase Price"] = round(age_sum_df["Price"]/purchase_agegrp_df["Purchase Count"],2).map("${:,.2f}".format)
purchase_agegrp_df["Total Purchase Value"] = age_sum_df["Price"]

purchase_agegrp_df["Normalized Totals"] =  round(purchase_agegrp_df["Total Purchase Value"]/purchase_agegrp_df["Total Count"],2).map("${:,.2f}".format)

purchase_agegrp_df["Total Purchase Value"] = purchase_agegrp_df["Total Purchase Value"].map("${:,.2f}".format)

purchase_agegrp_df = purchase_agegrp_df.drop(labels = ["Total Count","Percentage of Players"],axis = 1)

purchase_agegrp_df
```

```python
#Top Spenders

purchase_counts_df = pd.DataFrame(heroes_df.groupby("SN").count())

purchase_sum_df = pd.DataFrame(heroes_df.groupby("SN").sum())
top_spenders_df = purchase_sum_df
top_spenders_df = top_spenders_df.sort_values(by=["Price"], ascending=False)

top_spenders_df["Purchase Count"] = purchase_counts_df["Item ID"]
top_spenders_df["Average Purchase Price"] = round(top_spenders_df["Price"]/top_spenders_df["Purchase Count"],2).map("${:,.2f}".format)
top_spenders_df["Total Purchase Value"] = top_spenders_df["Price"].map("${:,.2f}".format)

top_spenders_df = top_spenders_df.drop(labels = ["Age","Item ID","Price"],axis = 1)

top_spenders_df.head(5)
```

```python
# MOST POPULAR ITEMS

items_count_df = pd.DataFrame(heroes_df.groupby(["Item ID", "Item Name"]).count())

items_sum_df = pd.DataFrame(heroes_df.groupby(["Item ID", "Item Name"]).sum())

items_count_df["Purchase Count"] = items_count_df["Age"]
items_count_df["Item Price"] = round(items_sum_df["Price"]/items_count_df["Age"],2)
items_count_df["Total Purchase Value"] = items_sum_df["Price"]


items_count_df = items_count_df.sort_values(by = ["Purchase Count","Total Purchase Value"],ascending = False)

popular_items_df = items_count_df.drop(labels = ["Age","Age Group","Gender","Price","SN"],axis = 1)

popular_items_df["Item Price"] = popular_items_df["Item Price"].map("${:,.2f}".format)
popular_items_df["Total Purchase Value"] = popular_items_df["Total Purchase Value"].map("${:,.2f}".format)


popular_items_df.head()
```

```python
#MOST PROFITABLE ITEMS

items_count_df = items_count_df.sort_values(by = ["Total Purchase Value"],ascending = False)

profitable_items_df = items_count_df.drop(labels = ["Age","Age Group","Gender","Price","SN"],axis = 1)

profitable_items_df["Item Price"] = profitable_items_df["Item Price"].map("${:,.2f}".format)
profitable_items_df["Total Purchase Value"] = profitable_items_df["Total Purchase Value"].map("${:,.2f}".format)

profitable_items_df.head()
```
